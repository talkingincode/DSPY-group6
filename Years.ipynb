```python
import pandas as pd
df = pd.read_csv('ufo-sightings-transformed.csv')
# print(df.shape)
print(df['Year'].min())
print(df['Year'].max())
df.head()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# 下載停用詞庫
nltk.download('stopwords')
nltk.download('punkt')
```

```python
# 將Description欄位轉換為文本字符串
# 依照每十年分類
text1 = "" # 1906-1915
text2 = "" # 1916-1925
text3 = "" # 1926-1935
text4 = "" # 1936-1945
text5 = "" # 1946-1955
text6 = "" # 1956-1965
text7 = "" # 1966-1975
text8 = "" # 1976-1985
text9 = "" # 1986-1995
text10 = "" # 1996-2005
text11 = "" # 2006-2015


for _, row in df.iterrows():
    if 1906 <= row['Year'] < 1916:
        if type(row['Description']) == float:
            text1 += str(row['Description'])
        else:
            text1 += row['Description']
        text1 += " "
    elif 1916 <= row['Year'] < 1926:
        if type(row['Description']) == float:
            text2 += str(row['Description'])
        else:
            text2 += row['Description']
        text2 += " "
    elif 1926 <= row['Year'] < 1936:
        if type(row['Description']) == float:
            text3 += str(row['Description'])
        else:
            text3 += row['Description']
        text3 += " "
    elif 1936 <= row['Year'] < 1946:
        if type(row['Description']) == float:
            text4 += str(row['Description'])
        else:
            text4 += row['Description']
        text4 += " "
    elif 1946 <= row['Year'] < 1956:
        if type(row['Description']) == float:
            text5 += str(row['Description'])
        else:
            text5 += row['Description']
        text5 += " "
    elif 1956 <= row['Year'] < 1966:
        if type(row['Description']) == float:
            text6 += str(row['Description'])
        else:
            text6 += row['Description']
        text6 += " "
    elif 1966 <= row['Year'] < 1976:
        if type(row['Description']) == float:
            text7 += str(row['Description'])
        else:
            text7 += row['Description']
        text7 += " "
    elif 1976 <= row['Year'] < 1986:
        if type(row['Description']) == float:
            text8 += str(row['Description'])
        else:
            text8 += row['Description']
        text8 += " "
    elif 1986 <= row['Year'] < 1996:
        if type(row['Description']) == float:
            text9 += str(row['Description'])
        else:
            text9 += row['Description']
        text9 += " "
    elif 1996 <= row['Year'] < 2006:
        if type(row['Description']) == float:
            text10 += str(row['Description'])
        else:
            text10 += row['Description']
        text10 += " "
    elif 2006 <= row['Year'] < 2016:
        if type(row['Description']) == float:
            text11 += str(row['Description'])
        else:
            text11 += row['Description']
        text11 += " "
```

```python
# 指定文件名和路径
file_path = "./text10.txt"
# 使用 with open() 结构将字符串列表写入文本文件
with open(file_path, 'w') as file:
    for line in text10:
        file.write(line )

# 指定文件名和路径
file_path = "./text11.txt"
# 使用 with open() 结构将字符串列表写入文本文件
with open(file_path, 'w') as file:
    for line in text11:
        file.write(line )
```

```python
# 將文本字符串分詞
words1 = word_tokenize(text1)
words2 = word_tokenize(text2)
words3 = word_tokenize(text3)
words4 = word_tokenize(text4)
words5 = word_tokenize(text5)
words6 = word_tokenize(text6)
words7 = word_tokenize(text7)
words8 = word_tokenize(text8)
words9 = word_tokenize(text9)

with open("text10.txt", "r") as file:
    text10 = file.read()
words10 = word_tokenize(text10)

with open("text11.txt", "r") as file:
    text10 = file.read()
words11 = word_tokenize(text11)
```

```python
# 去除停用詞and'quot'
filtered_words1 = [word for word in words1 if word.lower() not in stopwords.words('english') and word.lower() != 'quot']
filtered_words2 = [word for word in words2 if word.lower() not in stopwords.words('english') and word.lower() != 'quot']
filtered_words3 = [word for word in words3 if word.lower() not in stopwords.words('english') and word.lower() != 'quot']
filtered_words4 = [word for word in words4 if word.lower() not in stopwords.words('english') and word.lower() != 'quot']
filtered_words5 = [word for word in words5 if word.lower() not in stopwords.words('english') and word.lower() != 'quot']
filtered_words6 = [word for word in words6 if word.lower() not in stopwords.words('english') and word.lower() != 'quot']
filtered_words7 = [word for word in words7 if word.lower() not in stopwords.words('english') and word.lower() != 'quot']
filtered_words8 = [word for word in words8 if word.lower() not in stopwords.words('english') and word.lower() != 'quot']
filtered_words9 = [word for word in words9 if word.lower() not in stopwords.words('english') and word.lower() != 'quot']
filtered_words10 = [word for word in words10 if word.lower() not in stopwords.words('english') and word.lower() != 'quot']
filtered_words11 = [word for word in words11 if word.lower() not in stopwords.words('english') and word.lower() != 'quot']
```
